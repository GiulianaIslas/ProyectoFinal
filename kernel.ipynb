## Dependencias

```python
import matplotlib.pyplot as plt
import numpy
from PIL import Image
from scipy import ndimage
```

Normalizar la imagen

```python
Is = Image.open('/content/como_piensan_los_perros_7169_orig.jpg');
I = Is.convert('L');
I = numpy.asarray(I);
I = I / 255.0;
```

Selecciona el kernel que va a cambiar la nitidez de la imagen

```python
k1 = numpy.array([[0,-1,0],[-1,5,-1],[0,-1,0]])
```

Hace la convolusion con el kernel que se ingresÃ³ 

```python
J1 = ndimage.convolve(I, k1, mode='constant', cval=0.0)
```

Mostrar 

```python

plt.figure(figsize = (20,20))

plt.subplot(3,3,1)
plt.imshow(Is)
plt.xlabel('Input Image')

plt.subplot(3,3,2)
plt.imshow(J1)
plt.xlabel('Sharpen filter')
```
