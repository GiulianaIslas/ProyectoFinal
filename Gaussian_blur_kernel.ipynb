## Dependencias

```
import matplotlib.pyplot as plt
import numpy
from PIL import Image
from scipy import ndimage
```

## Convolucion

Aqui se abre y normaliza la imagen

```
Is = Image.open('gato.png');
I = Is.convert('L');
I = numpy.asarray(I);
I = I / 255.0;
```

Seleccionar los kernels y aplicar la convolucion

```
def k1(sigma, size):
  A= numpy.zeros((size,size))
  for x in range(-size//2,size//2+1):
    for y in range (-size//2,size//2+1):
      A[x][y] = 1/(2*numpy.pi*sigma**2)*(numpy.exp(-(x**2 + y**2)/(2*sigma**2)))
  return A

# I * kn

J1 = ndimage.convolve(I, k1(15,15), mode='constant', cval=0.0)

plt.figure(figsize = (15,15))

plt.subplot(2,2,1)
plt.imshow(Is)
plt.xlabel('Input Image')

plt.subplot(2,2,2)
plt.imshow(J1)
plt.xlabel('Kernel Gaussian blur')


plt.grid(False)
plt.show()
```
